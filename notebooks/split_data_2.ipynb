```python
import os
import random
import shutil
```

```python
DATA_DIR = '/kaggle/working/vsl_224_subset' # training dataset
OUTPUT_DIR = "/kaggle/working/vsl_split"
os.makedirs(OUTPUT_DIR, exist_ok=True)

TRAIN_RATIO = 0.8
VALID_RATIO = 0.1
TEST_RATIO  = 0.1
```

```python
labels = sorted([d for d in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, d))])

for label in labels:
    label_dir = os.path.join(DATA_DIR, label)
    images = [f for f in os.listdir(label_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    # Shuffle
    random.shuffle(images)

    n = len(images)
    n_train = int(n * TRAIN_RATIO)
    n_valid = int(n * VALID_RATIO)
    n_test  = n - n_train - n_valid

    train_files = images[:n_train]
    valid_files = images[n_train:n_train+n_valid]
    test_files  = images[n_train+n_valid:]

    # Create folder structure
    for split in ["train", "valid", "test"]:
        os.makedirs(os.path.join(OUTPUT_DIR, split, label), exist_ok=True)

    # Copy files
    for fname in train_files:
        shutil.copy(os.path.join(label_dir, fname), os.path.join(OUTPUT_DIR, "train", label, fname))

    for fname in valid_files:
        shutil.copy(os.path.join(label_dir, fname), os.path.join(OUTPUT_DIR, "valid", label, fname))

    for fname in test_files:
        shutil.copy(os.path.join(label_dir, fname), os.path.join(OUTPUT_DIR, "test", label, fname))

    print(f"{label}: train={len(train_files)}, valid={len(valid_files)}, test={len(test_files)}")

print("\nDone! Split dataset saved to:", OUTPUT_DIR)
```

```python
SOURCE_DIR = "/kaggle/working/vsl_split"
ZIP_PATH = "/kaggle/working/vsl_split"

# Tạo file vsl_224_subset.zip trong /kaggle/working
shutil.make_archive(ZIP_PATH, 'zip', SOURCE_DIR)

print("Done, saved at:", ZIP_PATH + ".zip")
```
